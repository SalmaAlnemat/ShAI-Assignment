# Data Cleaning

```python
#impute missing values in 'BasePay'with the mean because it will not introduce bias into the analysis
df['BasePay']= df['BasePay'].fillna(df['BasePay'].mean())
df[['OvertimePay','OtherPay']]= df[['OvertimePay','OtherPay']].fillna(df[['OvertimePay','OtherPay']].mean())

null=df[['BasePay','OvertimePay','OtherPay']].isnull().sum()
null
```

```python
#Median imputation is often chosen when the data is skewed or has outliers, as it is less sensitive to extreme values compared to mean imputation.
df['Benefits']= df['Benefits'].fillna(df['Benefits'].median())

print(df[['Benefits']].isnull().sum())
```

```python
#imputation with a placeholder.
df['Notes']=df['Notes'].fillna('No Notes')
print(df[['Notes']].isnull().sum())

print(df['Notes'].head(10))
```

```python
df['Status']=df['Status'].fillna('Unknown')
print(df[['Status']].isnull().sum())

print(df['Status'].head(10))
```

# Simple Correlation Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Salaries.csv')

salary_column = 'TotalPay'
other_column = 'Benefits'

correlation_coefficient = df['TotalPay'].corr(df['Benefits'])

print(f'Correlation coefficient between {salary_column} and {other_column}: {correlation_coefficient}')


```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df[salary_column], y=df[other_column])
plt.title(f'Scatter Plot: {salary_column} vs {other_column}')
plt.xlabel(salary_column)
plt.ylabel(other_column)
plt.show()
```
